```python
import numpy as np
```

```python
nums = np.random.randint(low=230,high=890,size=1000)
```

```python
nums
```

```python
even = list()
prime=list()

for num in nums:
    rem = num % 2
    if rem == 0:
        even.append(num)
    else:
        for divisor in range(2,num):
            rem = num % divisor
            if rem == 0:
                break
            else:
                if divisor == (num-1):
                    prime.append(num)
                    break
                        
D={'even':even,'prime':prime}               
       
                    
```
